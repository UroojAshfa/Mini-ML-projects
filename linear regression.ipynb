```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canadaincome.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter( df.year, df['per capita income (US$)'], color="blue", marker='+')



```

```python
new_df = df.drop('per capita income',axis='columns')
new_df
```

```python
year = df['per capita income']
year

```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, year)
```

```python

reg.predict([[1900]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
