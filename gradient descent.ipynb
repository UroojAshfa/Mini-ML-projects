```python
import numpy as np
import matplotlib.pyplot as plt
   #gradient descent is an algorithm that finds best fit line for given training dataset
```

```python
def gradient_descent(x,y):
    mcurr = bcurr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.001
    
    for i in range(iterations):
        y_predicted = mcurr*x + bcurr
        md = -(2/n)*sum(x*(y-y_predicted))  # m = m - learning rate * /
        bd = -(2/n)*sum((y-y_predicted))   # b = b - learning rate * /b
        
        mcurr = mcurr - (learning_rate*md)
        bcurr = bcurr - (learning_rate*bd)
        print ("m = {}, b = {}, iteration{}".format(mcurr, bcurr, i))

x = np.array([1,2,3,4,5])
y= np.array([5,7,9,11,13])

gradient_descent(x,y)
```
